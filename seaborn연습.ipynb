```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x="class",data=titanic)
plt.title("타이타닉호에 각 클래스별 승객수")
```

```python
tips = sns.load_dataset("tips")
sns.countplot(x="day",data=tips)
plt.show()
#  요일별 팁준 횟수가 카운트 된 것임. 
```

```python
tips.head(10)
```

```python
#요일별 식사 금액 
sns.boxplot(x="day",y ="total_bill",data=tips)
```

```python
#요일별 식사 금액, 흡연자와 비흡연자를 비교
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips)
plt.show()
```

```python
#요일별 식사 금액, 성별을 비교 
sns.violinplot(x="day",y="total_bill",hue="sex",data=tips)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill",y = "tip", data=tips)
plt.show()
```

```python
#식사금액과 팁의 흡연자 비흡연자 기준, 색 변경 
sns.lmplot(x="total_bill",y="tip", hue="smoker",data=tips,palette ="Set3", size=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
fgp = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
#YlGnBu 파란색과 노란색 계열 
#RdYlGn_r
sns.heatmap(fgp, cmap="YlGnBu",annot = True, fmt = 'd') #정수는 formating = %d 로 바꿔준다.
plt.show()
```
