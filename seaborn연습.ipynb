```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x="class",data=titanic)
plt.title("   ")
```

```python
tips = sns.load_dataset("tips")
sns.countplot(x="day",data=tips)
plt.show()
#       . 
```

```python
tips.head(10)
```

```python
#   
sns.boxplot(x="day",y ="total_bill",data=tips)
```

```python
#  ,   
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips)
plt.show()
```

```python
#  ,   
sns.violinplot(x="day",y="total_bill",hue="sex",data=tips)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill",y = "tip", data=tips)
plt.show()
```

```python
#    ,   
sns.lmplot(x="total_bill",y="tip", hue="smoker",data=tips,palette ="Set3", size=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
# pivot(, , )
fgp = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
#YlGnBu    
#RdYlGn_r
sns.heatmap(fgp, cmap="YlGnBu",annot = True, fmt = 'd') # formating = %d  .
plt.show()
```

```python
#stripplot      
#jitter=True       . 
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
tit=titanic.pivot_table(index=["sex"], columns=["class"], aggfunc='size')
sns.heatmap(tit, annot=True, fmt='d', linewidths=20)
plt.show()
```

```python
sns.set_style("ticks")
iris = sns.load_dataset("iris")
iris
sns.pairplot(iris, hue="species")
plt.show()
```
