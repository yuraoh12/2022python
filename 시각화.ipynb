```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm 
import matplotlib
fig = plt.figure(constrained_layout = True)
#폰트 지정
font_location = "C:\Windows\Fonts\Arial.TTF"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = fontname)
#add_subplot(행의갯수, 열의갯수, 위치값)
#그래프의 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,3,3)
#원하는 위치에 그래프를 그림
x = range(0, 10)
y = range(5, 15)
ax1.plot(x, y, 'ro') 
c = range(5, 50)
d = range(5, 50)
ax2.plot(c, d,'ms')
# x축 이름, y축 이름 지정 
ax1.set_xlabel('time')
ax1.set_ylabel('qty')
ax2.set_xlabel('time')
ax2.set_ylabel('qty')
plt.show()
```

```python
import matplotlib.pyplot as plt 
# 막대그래프
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]
labels = ['english', 'math']
colors = ['red', 'blue']
#bar() 막대그래프를 그리는 함수 
ax.bar(x, y, tick_label = labels, color = colors)
for i, v in enumerate(y):
    ax.text(v+2, i+1.05,str(v), color = 'red', fontweight='bold')

plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3, 4]
y = [1, 2]
ax.barh(x, y, tick_label = labels)
plt.show
```

```python
import numpy as np
x = [1, 2]
y1, y2, y3 = [1,2], [2,5], [3,7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2 
ax.bar(x,y1,width=w, label ='english')
ax.bar(np.array(x)+w,y2,width=w, label = 'math')
ax.bar(np.array(x)+w*2,y3,width=w, label = 'korean')
ax.legend()
plt.show 
```

```python
#plot 차트 만들기
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['english', 'math', 'korean']
sizes = [45,35,20]
ax.pie(sizes, labels=label,autopct='%1.1f%%',shadow=True)
#차트 제목 설정 
ax.set_title('우리반 영어, 수학, 국어, 점수 차트')
plt.show()
```

```python
#box plot 그리기 
plt.style.use('ggplot')
fig=plt.figure()
ax = fig.add_subplot(111)
x = [1,2,3,3,11,20]
ax.boxplot(x)
plt.show()
```

```python
#여러개의 boxplot 만들기
#box plot 그리기
plt.style.use('ggplot')
fig=plt.figure()
ax=fig.add_subplot(111)
x=[[1,2,3,3,11,20],[1,2,9,10,19,13]]
label = ['english', 'math']
ax.boxplot(x, labels = label)
plt.show()
```
